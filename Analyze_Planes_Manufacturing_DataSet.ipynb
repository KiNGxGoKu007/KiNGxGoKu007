```python
import pandas as pd
import numpy as np
```

```python
path="/content/drive/MyDrive/DataSetsAnalysis/planes.csv"
planes_df=pd.read_csv(path)
planes_df.head()
```

```python
planes_df['manufacturer'].unique()
```

```python
planes_df['engine'].unique()
```

```python
# Pandas solution
unique_values = planes_df['engine'].unique()
task1_my = pd.DataFrame(np.array(unique_values).reshape(-1, 1),
                        columns=['engine'])
task1_my
```

```python
# Pandas solution
task3_my = pd.DataFrame(planes_df.groupby(['engine']).size().reset_index(name="COUNT(*)"))
task3_my
```

```python
planes_df['engine'].value_counts()
```

```python
# Pandas solution
task4_my = planes_df.groupby(['engine', 'type']).size().reset_index(name='COUNT(*)')
task4_my
```

```python
# Pandas solution
task5_my = planes_df.groupby(['engine', 'manufacturer'])['year'].agg(['min', 'mean', 'max']).reset_index()
task5_my.head()
task5_my.rename(columns={'min': 'MIN(year)', 'mean': 'AVG(year)',
                         'max': 'MAX(year)'}, inplace=True)
task5_my
```

```python
# Pandas solution
task6_my = planes_df[planes_df['speed'].notnull()]
task6_my = task6_my.reset_index(drop=True)
task6_my
```

```python
# Pandas solution
task7_my = planes_df[(planes_df['year']>=2012) & (planes_df['seats']>150) & (planes_df['seats']<190)]['tailnum']
task7_my = task7_my.reset_index(drop=True)
task7_my = pd.DataFrame(task7_my)
task7_my
```

```python
# Pandas solution
list_manufacturer = ["BOEING", "AIRBUS", "EMBRAER"]
task8_my = planes_df[(planes_df['manufacturer'].isin(list_manufacturer)) & (planes_df['seats']>390)]
task8_my = task8_my[['tailnum','manufacturer','seats']]
task8_my.reset_index(drop=True,inplace=True)
task8_my
```

```python
# Pandas solution
task9_my = planes_df.groupby('manufacturer')['manufacturer'].count()
task9_my = pd.DataFrame(task9_my)
task9_my.rename(columns={'manufacturer':'COUNT(*)'}, inplace=True)
task9_my = task9_my[task9_my['COUNT(*)']>10]
task9_my.reset_index(inplace=True)
task9_my
```

```python
# Pandas solution
task10_my = planes_df.groupby('manufacturer')['manufacturer'].count()
task10_my = pd.DataFrame(task10_my)
task10_my.rename(columns={'manufacturer':'howmany'}, inplace=True)
task10_my.reset_index(inplace=True)
task10_my = task10_my.sort_values(by='howmany',ascending=False).head(5)
task10_my
```
