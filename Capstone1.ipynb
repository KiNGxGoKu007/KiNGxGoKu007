```python
female_weights = np.random.normal(loc=60, scale=10, size=100)
male_weights = np.random.normal(loc=70, scale=12, size=100)

fig, axs = plt.subplots(2, 1, figsize=(8, 6))

axs[0].hist(female_weights, bins=20, color='Green', alpha=0.7, label='Female')
axs[1].hist(male_weights, bins=20, color='Red', alpha=0.7, label='Male')

# Set limits for X-axis
x_min = min(min(female_weights), min(male_weights))
x_max = max(max(female_weights), max(male_weights))
axs[0].set_xlim(x_min, x_max)
axs[1].set_xlim(x_min, x_max)

axs[0].set_ylabel('Frequency')
axs[1].set_ylabel('Frequency')
axs[1].set_xlabel('Weight')
axs[0].set_title('Female Weight Distribution')
axs[1].set_title('Male Weight Distribution')

axs[0].legend()
axs[1].legend()

# Visualisation of the Graph
plt.tight_layout()
plt.show()
```

# **Advantages and Disadvantages**
**Advantages of BMI**

Advantages:

1. It's a good way to gauge whether your weight is in healthy proportion to your height.
2. Another key advantage of using BMI for determining weight-related health risk is that the measures needed for its calculation — height and weight — can be taken with high accuracy when performed using standardized procedures.

Disadvantages:

1. Limited Accuracy: BMI doesn’t consider factors like body composition, ethnicity, sex, race, or age.
2. Doesn’t Differentiate Fat Distribution: It treats all excess weight the same, regardless of where fat is stored.

**Advantages of Waist to Hip Ratio**

Advantages:

1. Better Predictor of Health: Considers fat distribution around the waist, hips, and buttocks.
2. Area-Specific: Targets the most problematic fat region—the waist.

Disadvantages:

1. Requires Measurement of Both Waist and Hip: Slightly more involved than BMI.
2. Cut-Offs Vary by Gender: Different thresholds for men and women.

**Advantages of Waist to Height Ratio**

Advantages:

1. Properly Reflects Central Obesity: It directly measures abdominal fat, a key indicator of central obesity. This type of fat deposit is linked to higher rates of metabolic syndrome, type 2 diabetes, heart disease, and other obesity-related conditions.
2. Consistent Across Ethnicities and Ages: It maintains a stable relationship with health outcomes across various populations, making it a universally valid metric. This consistency improves the accuracy of obesity diagnosis and health assessments.
3. Simple and Practical to Use: Measuring and calculating it is straightforward. its value of 0.5 or higher indicates an increased risk of obesity-related health problems12.

Disadvantages:

1. Measurement Accuracy
2. Limited aplicability
3. Not suitable for childerens
